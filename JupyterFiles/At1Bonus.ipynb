```python
import matplotlib.pyplot as plt
from matplotlib.patches import Polygon
from mpl_toolkits.basemap import Basemap
from numpy import arange
from matplotlib import pyplot as plt
from matplotlib import cm
from matplotlib.colors import LinearSegmentedColormap
from matplotlib.colors import rgb2hex
import matplotlib
import matplotlib.cm as cm
```

```python
#determine range to print based on min, max lat and long of the data\n",
margin = .2 # buffer to add to the range\n",
lat_min = -44 - margin
lat_max = 5 + margin
long_min = -76 - margin
long_max = -34 + margin

map = Basemap(llcrnrlon=long_min,
    llcrnrlat=lat_min,
    urcrnrlon=long_max,
    urcrnrlat=lat_max,
    lat_0=(lat_max - lat_min)/2,
    lon_0=(long_max-long_min)/2,
    projection='merc',
    resolution = 'h',
    area_thresh=10000
    )

```

```python
# This file will be generated by Irradiation_Data_Generator.py
file = open('output.txt', 'r+')
```

```python
lat = []
lon = []
weight = []
data_test = []
content = file.readlines()
for line in content:
    splited_line = line.split(';')
    lat.append(float(splited_line[0]))
    lon.append(float(splited_line[1]))
    temp1 = splited_line[2].strip('\n')
    temp2 = temp1.replace("'", "")    
    weight.append(float(temp2))
    
```

```python
minima = min(weight)
maxima = max(weight)
```

```python
norm = matplotlib.colors.Normalize(vmin=minima, vmax=maxima, clip=True)
mapper = cm.ScalarMappable(norm=norm, cmap=cm.hot)
```

```python
i = 0
for t in weight:
    #print(t)
    lons, lats = map(lon[i], lat[i])
    CS1 = map.scatter(lons, lats, marker = 'o', color=rgb2hex(mapper.to_rgba(t)), zorder=5)
    i = i+1
# adding colorbar:
mapper.set_array([]) # can be an empty list
cb = map.colorbar(mapper,"bottom", size="5%", pad="2%")
```
